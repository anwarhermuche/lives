### De todos os bilionários, quantos são CEOs?

```python
# Criando uma lista com valroes 1 para 'ceo' e 0 para não 'ceo' ou nulo
contain_ceo = np.array([0 if type(titulo) != str else 1 if 'ceo' in titulo.lower() else 0 for titulo in df.title.values])
```

```python
porcentagem_de_ceos = contain_ceo.mean()*100
```

```python
print(f"Temos {porcentagem_de_ceos:.2f}% de CEO's no dataset.")
```

Consideramos como sendo cargos de CEO todos os títulos que continham a palavra 'CEO' nele. Entretanto, temos 87.16% dos dados dessa coluna (title) nulos e isso impacta negativamente a nossa análise. 

Lembrando que a porcentagem foi calculada como sendo a média da lista 'contain_ceo'. Considerei como sendo '1', ou seja, "é um CEO", todos os títulos que continham CEO. Todo o resto (valores nulos e títulos que não contêm 'ceo') considerei como 0.

Abaixo, irei calcular a mesma porcentagem, porém excluindo os valores nulos.

```python
contain_ceo_novo = np.array([-1 if type(titulo) != str else 1 if 'ceo' in titulo.lower() else 0 for titulo in df.title.values])
```

```python
porcentagem_de_ceos_novo = contain_ceo_novo[contain_ceo_novo != -1].mean()*100
```

```python
print(f"Temos {porcentagem_de_ceos_novo:.2f}% de CEO's no dataset cujo título é não nulo.")
```

A partir disso, vemos que para todos os cargos que são não nulos, 27% dos bilionários desempenham o papel de CEO.

### Quais são as 10 indústrias com mais bilionários?

```python
# Código para responder a nossa pergunta
df.groupby('industries').count()[['rank']].\
            reset_index().\
            sort_values(by = 'rank', ascending = False).\
            rename(columns = {'rank': 'quantity'}).\
            reset_index(drop = True).\
            head(10)
```
